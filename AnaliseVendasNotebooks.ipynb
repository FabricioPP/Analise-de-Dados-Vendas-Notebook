```python
# prompt: importe pandas, matplotlib, seaborn e as bibliotecas para analises de dados

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/MOCK_DATA.csv', delimiter=',', encoding='1252')

df.head()
```

```python
df.shape
```

```python
# prompt: rotinas para ver se tem nulo, duplicados e ver tipos das colunas coloque para imprimir cada um formatado

print(df.isnull().sum())

df.duplicated().sum()

df.dtypes
```

```python
df['total_desconto'] = df['desconto'] * df['quantidade']
df['total_vendas'] = (df['valor_produto'] * df['quantidade']) - df['total_desconto']

df.head()
```

```python
# prompt: transformar coluna data_venda em datatime

df['data_venda'] = pd.to_datetime(df['data_venda'])
df.head()
```

```python
# prompt: criar coluna para ano e mes baseado na data da venda

df['ano'] = df['data_venda'].dt.year
df['mes'] = df['data_venda'].dt.month
```

```python
df.groupby('mes')['total_vendas'].sum()
```

```python


# Converter a coluna 'data_venda' para o tipo datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])

# Mapear os nomes dos meses em português
nomes_meses = {
    1: 'Janeiro',
    2: 'Fevereiro',
    3: 'Março',
    4: 'Abril',
    5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}

# Adicionar colunas de mês e ano usando o mapeamento
df['ano'] = df['data_venda'].dt.year
df['mes_nome'] = df['data_venda'].dt.month.map(nomes_meses)

# Agrupar e contar as vendas por mês
vendas_por_mes = df.groupby(['ano', 'mes', 'mes_nome'])['total_vendas'].sum().reset_index(name='quantidade_vendas')

# Agrupar e contar as vendas por ano
vendas_por_ano = df.groupby('ano').size().reset_index(name='quantidade_vendas')

# Exibir os resultados
print("Vendas por mês:\n", vendas_por_mes)
print("\nVendas por ano:\n", vendas_por_ano)
```

```python
import matplotlib.ticker as mticker

# Função para formatar valores em milhões
def milhoes_formatter(x, pos):
    return f'R${x/1e6:.1f}M'

# Gráfico de Barras
vendas_mes_sorted = vendas_por_mes.sort_values('quantidade_vendas', ascending=True)
plt.figure(figsize=(8, 5))
plt.barh(vendas_mes_sorted['mes_nome'], vendas_mes_sorted['quantidade_vendas'])
plt.xlabel('Total de Vendas (Milhões de Reais)')
plt.ylabel('Mês')
plt.title('Total de Vendas por Mês')

# Formatando o eixo x para mostrar os valores em milhões
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(milhoes_formatter))
plt.show()

# Gráfico de Linhas
plt.figure(figsize=(8, 5))
plt.plot(vendas_por_mes['mes_nome'], vendas_por_mes['quantidade_vendas'], marker='o')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas (Milhões de Reais)')
plt.title('Tendência de Vendas por Mês')
plt.xticks(rotation=30)
plt.ylim(0, max(vendas_por_mes['quantidade_vendas'])*1.8)
plt.grid(True)

# Formatando o eixo y para mostrar os valores em milhões
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(milhoes_formatter))
plt.show()
```

```python
# Calcular o total de vendas por mês
total_vendas_por_mes = df.groupby('mes_nome')['total_vendas'].sum()

# Converter a coluna 'mes_nome' de volta para string
df['mes_nome'] = df['mes_nome'].astype(str)

# Calcular a porcentagem de vendas para cada modelo em relação ao total de vendas por mês
df['porcentagem'] = df['total_vendas'] / df['mes_nome'].map(total_vendas_por_mes) * 100

# Definir a ordem dos meses
ordem_meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Converter a coluna 'mes_nome' para uma categoria com a ordem definida
df['mes_nome'] = pd.Categorical(df['mes_nome'], categories=ordem_meses, ordered=True)


# Tratar duplicatas, calculando a média das porcentagens para cada modelo em cada mês
df_agrupado = df.groupby(['mes_nome', 'modelo'])['porcentagem'].sum().reset_index()

# Criar um gráfico de barras empilhadas
plt.figure(figsize=(12, 8))
cores = sns.color_palette('Set3', n_colors=len(df['modelo'].unique()))
ax = df_agrupado.pivot(index='mes_nome', columns='modelo', values='porcentagem').plot(kind='bar', stacked=True, color=cores, width=0.7)
plt.xlabel('Mês')
plt.ylabel('Porcentagem de Vendas')
plt.title('Porcentagem de Vendas por Modelo e Mês')
plt.legend(title='Modelo', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar rótulos às barras
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    percent_value = height.round(2)
    ax.annotate(f'{percent_value}%', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=8)

# Ajustar a distância entre as barras
plt.subplots_adjust(wspace=0.3)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Cria um regressão linear
model = LinearRegression()

# Fit the model to the data
model.fit(df_novo.index.values[:, np.newaxis], df_novo.values)

# Make predictions
predictions = model.predict(df_novo.index.values[:, np.newaxis])

# Plot the data and the predictions
plt.scatter(df_novo.index, df_novo.values)
plt.plot(df_novo.index, predictions, color='red')
plt.show()

# Evaluate the model
print('R-squared:', model.score(df_novo.index.values[:, np.newaxis], df_novo.values))
```
